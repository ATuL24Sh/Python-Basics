```python
#Assignment 9: Matrix Transposition
#Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

def Transpose(matrix):
  T = [ [matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
  return T

'''
matrix[0] gives the number of columns in the original matrix, which determines how many rows the transposed matrix will have.
By iterating over range(len(matrix[0])), you ensure that you cover all columns of the original matrix to form the rows of the transposed matrix.

matrix is used to get the number of rows (len(matrix)).
matrix[0] is used to get the number of columns (len(matrix[0])).
This approach ensures that you correctly handle the dimensions of the matrix when creating its transpose.
'''

matrix=[
    [1,2,3,1],
    [4,5,6,2],
    [7,8,9,3]
    ]

print("Original Matrix A : ")
for row in matrix :
  print(f"{row}")

Tmatrix=Transpose(matrix)

print("Transpose of Matrix A : ")
for row in Tmatrix :
  print(f"{row}")
```

```python
#Assignment 10: Flattening a Nested List
#Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

def flat(neslst):

  '''
  #without list comprehension
  flst=[]
  for item in neslst:
    for i in item :
      flst.append(i)
  '''
  #with list comprehension
  flst=[i for items in neslst for i in items ]
  return flst

neslst=[[1,2,3],['A','B','C'],[1.05,True,False]]
flaten=flat(neslst)
print("Original")
print(neslst)
print("Flattened")
print(flaten)
```

```python
#Assignment 12: List Zipping
#Create two lists of the same length. Use the zip function to combine these lists into a list of tuples and print the result.

'''
The zip function is used to combine elements from multiple iterables (like lists or tuples) into tuples.
The resulting tuples contain elements from the corresponding positions of each iterable.
-zip combines multiple iterables element-wise into tuples.
-Use * to unzip tuples back into separate lists.
-When iterables have different lengths, zip stops at the shortest one.

syntax :    zip(iterable1, iterable2, ...)
'''

lst121=['a','b','c']
lst122=[1,2,3]
combined=list(zip(lst121,lst122))
print(combined)

lst121a,lst122a=zip(*combined)
print(lst121a)
print(lst122a)
```

```python
#Assignment 14: List Rotation
#Write a function that rotates a list by n positions. Print the original and rotated lists.

def rot(lst,n):
  n = n % len(lst14)
  return lst[-n:] + lst[:-n]

lst14=[1,2,3,4]
new =rot(lst14,3)
print(lst14)
print(new)
```

```python
#Assignment 15: List Intersection
#Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

def inter(lst1,lst2):
  return [i for i in lst1 if i in lst2]

lst15a=[1,2,3,4,5]
lst15b=[4,5,6,7,8]
new =inter(lst15a,lst15b)
print(new)
```

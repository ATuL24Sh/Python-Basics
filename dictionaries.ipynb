```python
#Assignment 1: Creating and Accessing Dictionaries
#Create a dictionary with the first 10 positive integers as keys and their squares as values. Print the dictionary.

d1={i : i**2 for i in range(1,11)}
print(d1)
```

```python
#Assignment 2: Accessing Dictionary Elements
#Print the value of the key 5 and the keys of the dictionary created in Assignment 1.

print(d1.get(5))
print(d1.keys())
```

```python
#Assignment 3: Dictionary Methods
#Add a new key-value pair (11, 121) to the dictionary created in Assignment 1 and then remove the key-value pair with key 1. Print the modified dictionary.

d1[11]=121
del d1[1]
print(d1)
```

```python
#Assignment 4: Iterating Over Dictionaries
#Iterate over the dictionary created in Assignment 1 and print each key-value pair.

for key,val in d1.items() :
  print(f"{key} : {val}")
```

```python
#Assignment 5: Dictionary Comprehensions
#Create a new dictionary containing the cubes of the first 10 positive integers using a dictionary comprehension. Print the new dictionary.

d2 = { i : i**3 for i in range(1,11)}
print(d2)
```

```python
#Assignment 6: Merging Dictionaries
#Create two dictionaries: one with keys as the first 5 positive integers and values as their squares, and another with keys as the next 5 positive integers and values as their squares.
#Merge these dictionaries into a single dictionary and print it.

d3= { i : i**2 for i in range(1,6)}
d4= { i : i**2 for i in range(6,11)}

d5 = {**d3,**d4}
print(d5)
```

```python
#Assignment 7: Nested Dictionaries
#Create a nested dictionary representing a student with keys 'name', 'age', 'grades', where 'grades' is another dictionary with keys 'math', 'science', and 'english'. Print the nested dictionary.

d6={'name' : 'atul', 'age' : 25, 'grades' : {'math' : 'A', 'science' : 'B', 'english' : 'B' } }

for key,val in d6.items():
  print(f'{key} : {val}')

#print(d6)
```

```python
#Assignment 8: Dictionary of Lists
#Create a dictionary where the keys are the first 5 positive integers and the values are lists containing the first 5 multiples of the key. Print the dictionary.

d7={i : [i*j for j in range(1,6)] for i in range(1,6)}

for key,val in d7.items():
  print(f'{key} : {val}')
#print(d7)
```

```python
#Assignment 9: Dictionary of Tuples
#Create a dictionary where the keys are the first 5 positive integers and the values are tuples containing the key and its square. Print the dictionary.

d8= { i : (i,i**2) for i in range(1,6) }

for k,v in d8.items():
  print(f'{k}:{v}')
```

```python
#Assignment 10: Dictionary and List Conversion
#Create a dictionary with the first 5 positive integers as keys and their squares as values. Convert the dictionary to a list of tuples and print it.

d9= {i : i**2 for i in range(1,6)}
print(f"dictionary : {d9}")
list_dict=list(d9.items())
print(list_dict)
```

```python
#Assignment 11: Dictionary Filtering
#Create a dictionary with the first 10 positive integers as keys and their squares as values.
#Create a new dictionary containing only the key-value pairs where the key is even. Print the new dictionary.

d10={i : i**2 for i in range(1,11)}
even_dict= {i:j for i,j in d10.items() if i%2==0}
print(even_dict)
```

```python
#Assignment 12: Dictionary Key and Value Transformation
#Create a dictionary with the first 5 positive integers as keys and their squares as values. Create a new dictionary with keys and values swapped. Print the new dictionary.

d11={i:i**2 for i in range(1,6)}

swap_dict= { j:i for i,j in d11.items()}
print(swap_dict)
```

```python
#Assignment 14: Counting with Dictionaries
#Write a function that takes a string and returns a dictionary with the count of each character in the string. Print the dictionary.

def str2dict(s):
  count=dict()
  for i in s:
    count[i]= count.get(i,0) + 1
  return count

string="anagram"
str2dict(string)



"""
count.get(i, 0):

count is a dictionary that keeps track of the number of occurrences of each character.
count.get(i, 0) is a method call that tries to get the value associated with the key i in the dictionary count.
If the key i exists in the dictionary, count.get(i, 0) returns its current value.
If the key i does not exist in the dictionary, count.get(i, 0) returns 0 (the default value provided).
count.get(i, 0) + 1:

This expression takes the value returned by count.get(i, 0) (which is the current count of the character i, or 0 if it’s not yet in the dictionary) and adds 1 to it.
This effectively increments the count of the character i by 1.
count[i] = ...:

This assignment updates the dictionary count by setting the value for the key i to the result of count.get(i, 0) + 1.
In other words, it records the new count of the character i in the dictionary.
Example
Let’s walk through an example with a string "abcab":

Initial Dictionary: {} (empty)
Iteration 1: Character 'a'
count.get('a', 0) returns 0 (since 'a' is not in the dictionary yet).
count['a'] = 0 + 1 sets the count of 'a' to 1.
Dictionary: {'a': 1}
Iteration 2: Character 'b'
count.get('b', 0) returns 0 (since 'b' is not in the dictionary yet).
count['b'] = 0 + 1 sets the count of 'b' to 1.
Dictionary: {'a': 1, 'b': 1}
Iteration 3: Character 'c'
count.get('c', 0) returns 0 (since 'c' is not in the dictionary yet).
count['c'] = 0 + 1 sets the count of 'c' to 1.
Dictionary: {'a': 1, 'b': 1, 'c': 1}
Iteration 4: Character 'a' (again)
count.get('a', 0) returns 1 (since 'a' is already in the dictionary with a count of 1).
count['a'] = 1 + 1 updates the count of 'a' to 2.
Dictionary: {'a': 2, 'b': 1, 'c': 1}
Iteration 5: Character 'b' (again)
count.get('b', 0) returns 1 (since 'b' is already in the dictionary with a count of 1).
count['b'] = 1 + 1 updates the count of 'b' to 2.

"""
```

```python
#Assignment 15: Dictionary and JSON
#Create a dictionary representing a book with keys 'title', 'author', 'year', and 'genre'. Convert the dictionary to a JSON string and print it.

import json
books={
    'title' : 'To Kill a Mockingbird',
    'author': 'Harper Lee',
    'year' : 1960,
    'genre' :'Fiction'
}
bookjson=json.dumps(books)

"""
The json.dumps() function is used to serialize a Python dictionary into a JSON string.
This is useful for data interchange formats, storing data in JSON format, or sending data over a network.
"""
print(f" JSON string : {bookjson}")
```
